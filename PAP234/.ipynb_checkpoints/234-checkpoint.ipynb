1.1) Himmelslicht Fenster offen/zu
===

```python
## loading data
# fenster zu
fz_lamb, fz_int = np.genfromtxt('data/himmelmitfenster.txt', skip_header=17, skip_footer=1, unpack=True, dtype='str') ## comma decimal points instead of dots
fz_lamb = [float(i.replace(',','.')) for i in fz_lamb]
fz_int = [float(i.replace(',','.')) for i in fz_int]
#fenster offen
fo_lamb, fo_int = np.genfromtxt('data/himmel.txt', skip_header=17, skip_footer=1, unpack=True, dtype='str') ## comma decimal points instead of dots
fo_lamb = [float(i.replace(',','.')) for i in fo_lamb]
fo_int = [float(i.replace(',','.')) for i in fo_int]

## offset; adding an offset, so that no value is <0
#fenster zu
offsetz = np.abs(np.min(fz_int))
fz_int = fz_int + offsetz
#fenster offen
offseto = np.abs(np.min(fo_int))
fo_int = fo_int + offseto

## Glasabsorption
fo_int_cal = fo_int
fo_int_cal[fo_int_cal == 0] == 1
A = 1. - np.array(fz_int)/np.array(fo_int_cal)

##plotting
fig = plt.figure(figsize=(12,9))
fig.subplots_adjust(hspace=.25)
ax = fig.add_subplot(211)
ax.plot(fz_lamb, fz_int, 'b-', label='$Fenster\ zu,\ Offset:\ ' + str(offsetz) + '$')
ax.plot(fo_lamb, fo_int, 'r-', label='$Fenster\ offen,\ Offset:\ ' + str(offseto) + '$')
ax.set_xlabel('$Wellenlnge\ \mathrm{[nm]}$')
ax.set_ylabel('$Intensitt\ \mathrm{[Counts]}$')
ax.set_title('$Plot\ 1:\ Spektrum\ des\ Himmelslichtes\ mit/ohne\ Fensterglas$')
ax.grid(axis='both')
ax.legend(loc=0)

ax = fig.add_subplot(212)
ax.plot(fo_lamb, A/np.max(A), 'k--', label='$Glasabsorption$')
ax.set_xlim(100, max(fo_lamb)+20)
ax.set_ylim(-0.7, 1.1)
ax.set_xlabel('$Wellenlnge\ \mathrm{[nm]}$')
ax.set_ylabel('$normierte\ Absorption\ \mathrm{A}$')
ax.set_title('$Plot\ 2:\ Glasabsorption$')
ax.grid(axis='y')
ax.legend(loc=0)

#plt.savefig('Plot12.pdf', dpi=400, format='pdf')
```

3) Auswertung des Natriumspektrums
===

- Starke Linien:

```python
## loading data
# natrium stark nicht gesttigt
na_stark_lamb, na_stark_int = np.genfromtxt('data/Natriumdampfhoch.txt', 
                                            skip_header=17, skip_footer=1, unpack=True, dtype='str') ## comma decimal points instead of dots
na_stark_lamb = [float(i.replace(',','.')) for i in na_stark_lamb]
na_stark_int = [float(i.replace(',','.')) for i in na_stark_int]

## offset
offset_na_stark = np.abs(np.min(na_stark_int))
na_stark_int = na_stark_int + offset_na_stark

## find the wavelengths:
## find maxes
def find_max(data, threshold, upper_lim):
    maxes = []
    for i,j in enumerate(data[:-1]):
        if( data[i-1] < j and j > data[i+1] and j > threshold and j < upper_lim):
            maxes.append((i,j))
    return maxes

maxima = find_max(na_stark_int, 550, np.infty)
#print(len(maxima))
maxima = np.array(maxima).transpose()
index = list(np.array(maxima[0], dtype='int'))

## plotting
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.plot(na_stark_lamb, na_stark_int, 'r-', label='$starke\ Na-Linien,\ Offset:\ ' + str(offset_na_stark) + '$')
#ax.plot(np.array(na_stark_lamb)[index], np.array(na_stark_int)[index], 'b.')
ax.set_yscale('log')
ax.set_xlim(250, 850)
ax.set_ylim(100,100000)
ax.set_xlabel('$Wellenlnge\ \mathrm{[nm]}$')
ax.set_ylabel('$\mathrm{log}\ Intensitt\ \mathrm{[Counts]}$')
ax.set_title('$Plot\ 6:\ Starke\ Linien\ des\ Natriumspektrums$')
    
ax.grid()
ax.legend(loc=0)
#plt.xlim([300,340])
#plt.savefig('Plot6.pdf', dpi=400, format='pdf')
```

Bestimmung der Serienenergien und der l-abhngigen Korrekturfaktoren
====

```python
ERy = -13.605
hc = 1.2398e3
### E3p
E3p, dE3p = ERy/3**2-hc/819.03, hc/819.03**2
printResult("E_{3p}",E3p,error=dE3p,unit="eV")

## E3s
E3s, dE3s = E3p-hc/588.94, np.sqrt(dE3p**2+(hc/588.94**2))
printResult("E_{3s}",E3s,error=dE3s,unit="eV")

## Delta s
Ds, dDs = 3-np.sqrt(ERy/E3s), dE3s/2*np.sqrt(ERy/E3s**3)
printResult("\Delta s",Ds,error=dDs,unit="")
```

```python
### data
### theoretisch
# erste serie
m1t = np.linspace(3,12,10)
series1t = series_fit(m1t,ERy,0)

# zweite serie
m2t = np.linspace(4,9,6)
series2t = series_fit(m2t,ERy,Ds)

# erste serie
m1 = np.array([3, 4, 5, 6, 7, 9, 10])
series1 = np.array([819.21, 568.32, 498.19, 466.86, 450.45 ,433.29, 429.6])
series1_err = np.array([3, 5, 2, 1, 1, 1, 2])

#zweite serie
m2 = np.array([6,8,9])
series2 = np.array([515.3,465.2,438.9])
series2_err = np.array([3, 2, 2])
```

```python
### fitting
def series_fit(m, Ery, delta_d):
    ## for series 2 delta_d is delta_s
    return hc/(Ery/(m-delta_d)**2 - E3p)

# series1
popt_s1, pcov_s1 = curve_fit(series_fit, m1, series1, p0=[ERy, 1])   
popt_s1_err = np.sqrt(np.diag(pcov_s1))
print('series1:\n',popt_s1, popt_s1_err)
par_str1 = r'$E_{Ry}=-12.95\pm1.20\ \mathrm{[eV]}$' + '\n' + r'$E_{3p}=-3.03\pm0.02\ \mathrm{[eV]}$' + '\n' + r'$\vartriangle_d=0.08\pm0.12$'

# series2
popt_s2, pcov_s2 = curve_fit(series_fit, m2, series2, p0=[ERy, 1])   
popt_s2_err = np.sqrt(np.diag(pcov_s2))
print('series2:\n',popt_s2, popt_s2_err)
par_str2 = r'$E_{Ry}=-13.03\pm0.23\ \mathrm{[eV]}$' + '\n' + r'$E_{3p}=-3.03\pm0.05\ \mathrm{[eV]}$' + '\n' + r'$\vartriangle_s=1.42\pm0.12$'

### fitting with fixed E_Ry
def series_fit_fixed_Ery(m,delta_d):
    ## for series 2 delta_d is delta_s
    return hc/(ERy/((m-delta_d)**2) - E3p)

# series1_fixed Ery
popt_s1_fixed, pcov_s1_fixed = curve_fit(series_fit_fixed_Ery, m1, series1, p0=[ 0.05])   
popt_s1_fixed_err = np.sqrt(np.diag(pcov_s1_fixed))
print('series1_fixed:\n',popt_s1_fixed, popt_s1_fixed_err)
par_str1_fixed = r'$E_{3p}=-3.04\pm0.01\ \mathrm{[eV]}$' + '\n' + r'$\vartriangle_d=0.01\pm0.01$'

# series2_fixed Ery
popt_s2_fixed, pcov_s2_fixed = curve_fit(series_fit_fixed_Ery, m2, series2, p0=[ 1])   
popt_s2_fixed_err = np.sqrt(np.diag(pcov_s2_fixed))
print('series2_fixed:\n',popt_s2_fixed, popt_s2_fixed_err)
par_str2_fixed = r'$E_{3p}=-3.039\pm0.001\ \mathrm{[eV]}$' + '\n' + r'$\vartriangle_s=1.357\pm0.003$'
```

```python
### plotting
fig = plt.figure(figsize=(12,9))
fig.subplots_adjust(hspace=.25)

# erste serie
params_plot7 = r'Fit 1:' + '\n' + par_str1 + '\n' + 'Fit 2:' + '\n' + par_str1_fixed
ax = fig.add_subplot(211)
ax.errorbar(m1, series1, yerr=series1_err, fmt='r.', label='$Messwerte\ der\ Wellenlngen$')

linspc1 = np.linspace(m1[:1],m1[-1:],100)

ax.plot(linspc1, series_fit(linspc1, *popt_s1), 'k--', label='$Fit\ 1:\ \\frac{1.2398\cdot10^3}{E_{Ry}/(m-\\vartriangle_d)^2-E_{3p}}$')
ax.plot(linspc1, series_fit_fixed_Ery(linspc1, *popt_s1_fixed), 'b-', 
        label='$Fit\ 2:\ \\frac{1.2398\cdot10^3}{-13.605/(m-\\vartriangle_d)^2-E_{3p}}$')
ax.set_xlim(2,12)
ax.text(7,610, params_plot7)
ax.set_title('$Plot\ 9:\ Gemessene\ Wellenlngen\ der\ 1ten\ Nebenserie\ ber\ Quantenzahl\ m\ mit\ Fit(s)$')
ax.set_xlabel('$Quantenzahl\ \mathrm{m}$')
ax.set_ylabel('$gemessene\ Wellenlnge\ \lambda\ \mathrm{[nm]}$')
ax.grid()
ax.legend(fontsize=12)

# zweite serie
params_plot8 = r'Fit 1:' + '\n' + par_str2 + '\n' + 'Fit 2:' + '\n' + par_str2_fixed
ax = fig.add_subplot(212)
ax.errorbar(m2, series2, yerr=series2_err, fmt='r.', label='$Messwerte\ der\ Wellenlngen$')

linspc2 = np.linspace(m2[:1],m2[-1:],100)

ax.plot(linspc2, series_fit(linspc2, *popt_s2), 'k--', label='$Fit\ 1:\ \\frac{1.2398\cdot10^3}{E_{Ry}/(m-\\vartriangle_s)^2-E_{3p}}$')
ax.plot(linspc2, series_fit_fixed_Ery(linspc2, *popt_s2_fixed), 'b-', 
        label='$Fit\ 2:\ \\frac{1.2398\cdot10^3}{-13.605/(m-\\vartriangle_s)^2-E_{3p}}$')
ax.set_xlim(5.5,9.5)
ax.text(7.5, 475, params_plot8)
ax.set_title('$Plot\ 10:\ Gemessene\ Wellenlngen\ der\ 2ten\ Nebenserie\ ber\ Quantenzahl\ m\ mit\ Fit(s)$')
ax.set_xlabel('$Quantenzahl\ \mathrm{m}$')
ax.set_ylabel('$gemessene\ Wellenlnge\ \lambda\ \mathrm{[nm]}$')
ax.grid()
ax.legend(fontsize=12)
#plt.savefig('plot910.pdf', dpi = 400, format='pdf')
```
