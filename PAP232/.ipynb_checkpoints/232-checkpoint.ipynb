```python
import matplotlib 
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from helper import *
from scipy.optimize import curve_fit
from pandas import DataFrame as df
%matplotlib inline
```

## Versuchsteil 1: Wellenlänge des Lasers 

```python
data1 = np.genfromtxt("data/teil1.dat",skip_header=1)
#print(data1)
sa = data1[:,0]
sb = data1[:,1]
n = data1[:,2]
ds = .0000005

lamb = 2*(sb-sa)/n
lambq = np.mean(lamb)
slamb = np.sqrt((np.sum((lambq-lamb)**2))/20)

dlamb = np.sqrt(2)*(ds*2/n)
dlambqsys = np.sqrt(np.sum((dlamb)**2))/5
dlambq = np.sqrt(dlambqsys**2+slamb**2)

df(np.array([lamb,dlamb]).T,
  columns=["$\lambda$","$\Delta\lambda$"])
```

```python
printResult("\overline\lambda",lambq*10**9,0,"nm")
printResult("\sigma_{\overline\lambda}",slamb,0)
printResult("\Delta\overline\lambda_{sys}",dlambqsys,0)
printResult("\Delta\overline\lambda",dlambq*10**9,0,"nm")
```

## Versuchsteil 2: Brechungsindex von Luft

```python
T0 = 273.15
T,dT = T0 + 23.5, 2
a,da = 5*10**-2,5*10**-5
p0 = 760
lamb,dlamb = 532*10**-9,10**-9
dp = 5

data2 = np.array([[-740,-665,-595],[-700,-610,-585],[-700,-640,-565]])
data22 = np.array([0,5,10])

curve1,popt1,pcov1 = fit_lin(data2[0],data22)
curve2,popt2,pcov2 = fit_lin(data2[1],data22)
curve3,popt3,pcov3 = fit_lin(data2[2],data22)

plt.errorbar(data2[0],data22,xerr=dp,marker=".",linestyle="none")
plt.errorbar(data2[1],data22,xerr=dp,marker=".",linestyle="none")
plt.errorbar(data2[2],data22,xerr=dp,marker=".",linestyle="none")

x = np.linspace(-800,-500,100)

plt.plot(x,popt1[0]*x+popt1[1],color="blue")
plt.plot(x,popt2[0]*x+popt2[1],color="red")
plt.plot(x,popt3[0]*x+popt3[1],color="green")

plt.title("m-p-Diagramm während Druckausgleich in der Küvette")
plt.xlabel("Druck [Torr]")
plt.ylabel("Anzahl Streifen")
plt.xlim((-745,-540))
plt.ylim((-1,11))
plt.show()
```

```python
printResult("(\Delta m /p)_1",popt1[0],np.sqrt(pcov1[0,0]),"1/Torr")
printResult("(\Delta m /p)_2",popt2[0],np.sqrt(pcov2[0,0]),"1/Torr")
printResult("(\Delta m /p)_3",popt3[0],np.sqrt(pcov3[0,0]),"1/Torr")
```

```python
bq = (popt1[0]+popt2[0]+popt3[0])/3
dbq = np.sqrt(pcov1[0,0]+pcov2[0,0]+pcov3[0,0])/3
printResult("\overline{\Delta m/p}",bq,dbq,"1/Torr")
```

```python
n0 = lamb/2/a*bq*p0*T/T0
dn0 = np.sqrt((n0/lamb*dlamb)**2
              +(n0/bq*dbq)**2
              +(n0/a*da)**2
              +(n0/T*dT)**2)
printResult("n_0",n0+1,dn0)
```

## Versuchsteil 3: Interferogramm und Weißlichtbereich einer LED

```python
data = np.genfromtxt("data/weisslicht.csv",delimiter=",",skip_header=2)

t = data[:,0]
Int = data[:,1]

plt.figure(figsize=(8,6))
plt.plot(t,Int)
plt.xlim(-.05,.07)
plt.xlabel('Zeit / '+'s')
plt.ylabel('Intensitaet / '+' b.E.')
plt.title('Interferogramm einer LED')
plt.show()
```

```python
peakind = signal.find_peaks_cwt(Int, np.arange(1,30),noise_perc=20)

plt.figure(figsize=(8,6))
plt.plot(t[peakind], Int[peakind],marker='o',linewidth=0)
plt.plot(t,Int, color='red', linewidth=1)
plt.xlabel('Zeit / '+'s')
plt.ylabel('Intensitaet / '+' b.E.')
plt.title('Interferogramm einer LED')
plt.xlim((-.05,.07))
plt.show()
```

```python
def fitFunc(t, a, mu, sig):
    return a/np.sqrt(2*np.pi)/sig*np.exp(-(t-mu)**2/(2*sig**2))

init_vals = [0.002, 0.032, 0.015]
fitParams, fitCovariances = curve_fit(fitFunc, t[peakind], Int[peakind],p0=init_vals)

x=np.linspace(-.01,0.05,100)

plt.plot(x, fitFunc(x, fitParams[0], fitParams[1], fitParams[2]))
plt.plot(t[peakind], Int[peakind],marker='*', linewidth=0)
plt.xlabel('Zeit / '+'s')
plt.ylabel('Intensitaet / '+' b.E.')
plt.title('Interferogramm einer LED (Gaußfit)')
plt.xlim((-.05,.06))
plt.show()
```

```python
sig,dsig = fitParams[2],np.sqrt(fitCovariances[2,2])

printResult("A",fitParams[0],np.sqrt(fitCovariances[0,0]))
printResult("\mu",fitParams[1],np.sqrt(fitCovariances[1,1]))
printResult("\sigma",sig,dsig,"s")
printResult("B = 2\sigma",sig*2,dsig*2,"s")
```

```python
v,dv = .1*10**-3,5*10**-6
dL = np.sqrt((dsig*2*v)**2+(dv*2*sig)**2)
printResult("L",2*2*sig*v*10**-6,dL*10**-6,"\mu m")
```

## Versuchsteil 3.2: Anpassung mit Lorentzkurve

```python
def lorentz(t, a, gamma, omega):
    return a/(2*np.pi)*gamma/((t-omega)**2+gamma**2/4)             

init_vals = [0.002, 0.02, 0.035]
fitParamsLorentz, fitCovariances = curve_fit(lorentz, t[peakind], Int[peakind],p0=init_vals)

x=np.linspace(-0.04,0.1,100) #x-Werte fuer die Fitfunktion 
plt.plot(x, lorentz(x, fitParamsLorentz[0], fitParamsLorentz[1], fitParamsLorentz[2]))
plt.plot(t[peakind], Int[peakind],marker='*', linewidth=0)
plt.xlabel('Zeit / '+'s')
plt.ylabel('Intensitaet / '+' b.E.')
plt.title('Interferogramm einer LED (Lorentzfit)')
plt.xlim((-.05,.07))
plt.show()
```

```python
printResult("A",fitParams[0],np.sqrt(fitCovariances[0,0]))
printResult("\gamma",fitParams[1],np.sqrt(fitCovariances[1,1]),"s")
printResult("\omega",fitParams[2],np.sqrt(fitCovariances[2,2]))
```

```python
x=np.linspace(-0.04,0.1,100) #x-Werte fuer die Fitfunktion 

fig = plt.figure()

l1, = plt.plot(x, lorentz(x, fitParamsLorentz[0], fitParamsLorentz[1], fitParamsLorentz[2]),label="Lorentz")
l2, = plt.plot(x, fitFunc(x, fitParams[0], fitParams[1], fitParams[2]),label="Gauß")
l3, = plt.plot(t[peakind], Int[peakind],marker='.', linewidth=0)
plt.legend()
plt.xlabel('Zeit / '+'s')
plt.ylabel('Intensitaet / '+' b.E.')
plt.title('Interferogramm einer LED (Vergleich Gauß- Lorentzfit)')
plt.xlim((-.05,.07))
plt.show()
```
